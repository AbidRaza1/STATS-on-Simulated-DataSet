## Stats for Simulated data set

```r
# Setting Seed

set.seed(1)
```

```r
# Creating Vector x of 100 observatiosn drawn from N(0,1) distribution

x = c(rnorm(100,0,1))
```

```r
# Creating Vector eps of 100 observatiosn drawn from N(0,0.25) distribution

eps = c(rnorm(100,0,sqrt(0.25)))
```

```r
# Creating vector y  from x and eps using following model

y = -1 + (0.5 * x) + eps
```

```r
# Plot of x and y

plot(x,y,col = 'blue')
```

```r
# Fitting a least squares linear model to predict y using x

leastsqrs = lm(y~x)
```

```r
# Summary of fit

summary(leastsqrs)
```

Findings:

1.Estimated Intercept, Coeff. and Actual Intercept, Coeff. are very close.

2.Relationship between x and y is not due to chance.

3.Model rejects the Null Hypothesis.



```r
#least squares line and population regression line

plot(x,y,col = 'blue')

abline(leastsqrs, col = 'red')

legend('topleft',legend= c('least squares regr.','population regr.'), col= c(2,3), lty=1,bty="n")


abline(-1,0.5, col = 'green')
```

         The above figure represents least square regression line(red) and population regression line(green)

```r
# Fitting Polynomial regresion model

pl = lm(y~x+I(x^2))

# Summary

summary(pl)
```

```r
# Comparing two plots and finding anamolies

par(mfrow = c(2,2))

a = plot(predict(leastsqrs),residuals(leastsqrs),col = 'orange')

abline(0,0)

 
legend('topleft','Linear fit')

plot(predict(pl),residuals(pl), col ='green')

legend('topleft','Quadratic fit', bty = "o")

abline(0,0)

```

Findings:

1-Polynomial fit does not really change the fit 

2-Quadratic term in equation seems to be insignificant
